### 신경망을 위한 지도 학습 (Supervised Learing)

<br>
<br>
다음은 지도 학습의 몇 가지 예입니다.

| 입력(x) | 출력(y) | 응용분야 |
|----------|-----------|-------------|
|집 특징들 | 가격 | 부동산 |
|광고, 사용자 정보 | 광고 클릭?(0,1) | 온라인 광고 |
|이미지 | 객체(1,...,1000) | 사진 태깅(tagging) |
|오디오 | 문자 받아적기 | 음성 인식 |
|영어 | 중국어 | 기계번역 |
|이미지, 레이다 정보| 다른 차들의 위치 | 자율 주행 |
