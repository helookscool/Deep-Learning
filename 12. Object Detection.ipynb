### Non-max suppression 
<img src="./images/non_max_04.png" width="170" align="left" > $\quad$
<br>

$\quad$ 각 출력 예측은 다음과 같습니다. $ \begin{bmatrix} p_c \\ b_x \\ b_y \\ b_h \\ b_w \end{bmatrix} $ <br>
<br><br>

* $ p_c \le 0.6 $ 인 모든 상자를 폐기하십시오. <br><br>
* 남은 상자가 있는 동안 : <br><br>
  - 가장 큰 $p_c$ 를 갖는 상자를 선택하여 예측으로 출력합니다.<br><br>
  - 이전 단계의 출력 상자와 $IoU \ge 0.5$인 나머지 상자는 폐기합니다.
