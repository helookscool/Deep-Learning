### 미니-배치 크기 선택하기
* mini-batch size $= m$ 인 경우 : 배치(batch) 경사하강 : $(X^{\{1\}},Y^{\{1\}}) = (X,Y)$
<br>
<br>
* mini-batch size $= 1$ 인 경우 : 확률적(scochastic) 경사하강 : $(X^{\{1\}},Y^{\{1\}}) = (x^{(1)},y^{(1)}), \cdots, (X^{\{m\}},Y^{\{m\}}) =(x^{(m)},y^{(m)}) $ Mini-Batch
<br>
<br>
* 실제 : 1 과 m 사이.
<br>

### 미니-배치 크기 선택하기
<br>

#### 훈련세트가 작은 경우 ($m \le 2000$) : 배치 경사하강을 사용합니다.
<br><br>

#### 전형적인 미니-배치 크기 $ \Rightarrow \quad $ 64, 128, 256, 512, or 1024
<br><br><br>

#### 미니-배치 $(X^{\{t\}},Y^{\{t\}})$ 가 CPU/GPU 메모리에 맞는지 확인할 필요가 있습니다.<br><br><br><br><br>

### 모멘텀을 사용한 경사하강 : 세부적 구현
<br><br>

On iteration t: <br><br>
 $\quad $ Compute dW, db on the current mini-batch <br><br>
    $\quad v_{dW} = \beta v_{dW} + (1-\beta) dW $ <br><br>
    $\quad v_{db} = \beta v_{db} + (1-\beta) db $ <br><br>
    $\quad W = W - \alpha v_{dW}, b = b - \alpha v_{db}$ <br>
<br><br>

Hyperparameters: $\alpha, \beta \quad \beta=0.9$
<br><br>
