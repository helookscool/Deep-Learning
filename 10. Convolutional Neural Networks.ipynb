### 왜 컨볼루션인가?

<br><img src="./images/why_convolution_01.PNG" width="700"><br><br>


**매개변수 공유** : 이미지의 한 부분에서 유용한 특징 감지기 (예: 수직 모서리 감지기)는 이미지의 다른 부분에서 유용 할 수 있습니다.
<br> 
<br>

**희박한(드문드문) 연결** : 각 층에서 각 출력 값은 적은 수의 입력에만 의존합니다.
<br>
<br>
<hr>
